```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv ('HyderabadResturants.csv')
```

```python
df1 = df.drop('links', axis ='columns') # or (['links'],axis ="columns")  --- & For deleting columns -axis = 'columns'
```

```python
df1
```

```python
df1.info()                                # it shows information of table
```

```python
df1.drop_duplicates (inplace = True)    # For dropping duplicates 
```

```python
df1                                            # no found duplicates
```

```python
df1['ratings'].unique()                    # For checking unique value ---,  & handle 'New' , '-'  cleaning
```

```python
def handelratings (value):                       
    if (value =='New'or value =='-'):                        # cleaning 'New' , '-'  -- jo value 'New' , '-' hai use nan kardo
        return np.nan                                                # ya float value kar do ,,,
    else:
        return float(value)

df1['ratings'] =df1['ratings'].apply(handelratings)

```

```python
df1['ratings'].unique()    
```

```python
df1['ratings'].fillna(df1['ratings'].mean(),inplace = True)  # We filled nul value of raitngs by ratings mean() value
```

```python
df1['ratings'].isnull().sum()    
```

```python
df1['ratings'].mean()
```

```python
import plotly.graph_objects as go

# Create a figure and subplot
fig = go.Figure()

# Assuming df1 is a DataFrame containing your data
# You can specify the data using 'x' and 'y' arguments in go.Bar
fig.add_trace(go.Bar(x=df1['cuisine'].value_counts().index, y=df1['cuisine'].value_counts(), marker_color="blue"))

# Update the layout to set the x-axis labels to be displayed vertically
fig.update_layout(xaxis_tickangle=-90)

# Set the layout title and labels
fig.update_layout(
    title="Cuisine Count",
    xaxis_title="Cuisine",
    yaxis_title="Count"
)

# Show the figure
fig.show()
```
